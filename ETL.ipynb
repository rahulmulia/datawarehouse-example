```python
def get_files(filepath):

    all_files=[]

    for root, dirs, files in os.walk(filepath):
        files=glob.glob(os.path.join(root,'*.json'))
        for file in files:
            all_files.append(os.path.abspath(file))
    return all_files     
```

```python
filepath=log_files[0]
```

```python
df = pd.read_json(filepath,lines=True)
df
```

```python
df = df[df['page'] == 'NextSong']
df.head()
```

```python
t= pd.to_datetime(df['ts'])
t.head()
```

```python
time_data= [(tt.value, tt.day ,tt.hour, tt.day, tt.month, tt.year, tt.weekday()) for tt in t]
column_label =('timestamp', 'day','hour','year','month','year','weekday')
```

```python
time_df= pd.DataFrame(data=time_data, columns=column_label)
time_df.head()
```

```python
for i, row in time_df.iterrows():
    cur.execute(time_table_insert, list(row))
```

```python
user_df= df[['userId', 'firstName', 'lastName', 'gender', 'level']]
```

```python
for i, row in user_df.iterrows():
    cur.execute(user_table_insert,list(row))
```

```python
for index, row in df.iterrows():
    cur.execute(song_select,(row.song,row.artist,row.length))
    results= cur.fetchone()
    
    if results:
        songid, artistid= results
    else:
        songid, artistid = None, None
        
    songplay_data= (index,row['ts'], row['userId'], row['level'], songid, artistid, row['sessionId'],
                     row['location'], row['userAgent'])
    cur.execute(songplay_table_insert,songplay_data)
```
